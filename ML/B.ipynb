```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
import pickle
import os
'''"bajra.xlsx", "barley.xlsx", "black gram.xlsx", "chilli red.xlsx", "copra.xlsx", 
    "gram_raw.xlsx", "green gram.xlsx", "groundnut.xlsx", "jowar.xlsx", "lentil.xlsx", 
    "maize.xlsx", "mustard.xlsx", "niger seed.xlsx", "onion.xlsx", "paddy.xlsx", 
    "pegeonn pea.xlsx",'''
# List of .xlsx files
files = [ "potato.xlsx", "ragi.xlsx", "safflower.xlsx", "sesamum.xlsx", 
    "soyabean.xlsx", "sugarcane.xlsx", "sunflower.xlsx", "tomato.xlsx", "wheat1.xlsx"
]

# Directory for saving models
model_dir = "models"
os.makedirs(model_dir, exist_ok=True)

# Iterate over each file
for file in files:
    try:
        # Load the dataset
        df = pd.read_excel(file)
        print(f"Processing file: {file}")

        # Preprocessing
        df['Price Date'] = pd.to_datetime(df['Price Date'])
        df['Day'] = df['Price Date'].dt.day
        df['Month'] = df['Price Date'].dt.month
        df['Year'] = df['Price Date'].dt.year

        # Encode categorical features
        label_encoders = {}
        for col in ['District Name', 'Commodity', 'Market Name']:
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col])
            label_encoders[col] = le

        # Ensure Min Price < Max Price
        df = df[df['Min Price (Rs./Quintal)'] <= df['Max Price (Rs./Quintal)']]

        # Define features and target variables
        X = df[['District Name', 'Commodity', 'Market Name', 'Day', 'Month', 'Year']]
        y_min = df['Min Price (Rs./Quintal)']
        y_max = df['Max Price (Rs./Quintal)']
        y_modal = df['Modal Price (Rs./Quintal)']

        # Split data into training and testing sets
        X_train, X_test, y_min_train, y_min_test = train_test_split(X, y_min, test_size=0.2, random_state=42)
        _, _, y_max_train, y_max_test = train_test_split(X, y_max, test_size=0.2, random_state=42)
        _, _, y_modal_train, y_modal_test = train_test_split(X, y_modal, test_size=0.2, random_state=42)

        # Train Random Forest models
        min_price_model = RandomForestRegressor(n_estimators=100, random_state=42)
        max_price_model = RandomForestRegressor(n_estimators=100, random_state=42)
        modal_price_model = RandomForestRegressor(n_estimators=100, random_state=42)

        min_price_model.fit(X_train, y_min_train)
        max_price_model.fit(X_train, y_max_train)
        modal_price_model.fit(X_train, y_modal_train)

        # Save models to .pkl files
        base_name = os.path.splitext(file)[0]  # Extract the base name from the file name
        with open(os.path.join(model_dir, f"{base_name}_min_price_model.pkl"), 'wb') as f:
            pickle.dump(min_price_model, f)
        with open(os.path.join(model_dir, f"{base_name}_max_price_model.pkl"), 'wb') as f: 
            pickle.dump(max_price_model, f)
        with open(os.path.join(model_dir, f"{base_name}_modal_price_model.pkl"), 'wb') as f:
            pickle.dump(modal_price_model, f)

        print(f"Models saved for {file}")

    except Exception as e:
        print(f"Error processing file {file}: {e}")
```

```python
import pickle
import numpy as np

# Mapping of commodity names to corresponding model base names
commodity_to_file_base = {
    "Bajra(Pearl Millet/Cumbu)": "bajra",
    "Barley (Jau)": "barley",
    "Black Gram (Urd Beans)(Whole)": "black gram",
    "Chili Red": "chilli red",
    "Copra": "copra",
    "Gram Raw(Chholia)": "gram_raw",
    "Green Gram (Moong)(Whole)": "green gram",
    "Groundnut": "groundnut",
    "Jowar(Sorghum)": "jowar",
    "Lentil (Masur)(Whole)": "lentil",
    "Maize": "maize",
    "Mustard": "mustard",
    "Niger Seed (Ramtil)": "niger seed",
    "Onion": "onion",
    "Paddy(Dhan)(Common)": "paddy",
    "Pegeon Pea (Arhar Fali)": "pegeon pea",
    "Potato": "potato",
    "Ragi (Finger Millet)": "ragi",
    "Safflower": "safflower",
    "Sesamum(Sesame,Gingelly,Til)": "sesamum",
    "Soyabean": "soyabean",
    "Sugarcane": "sugarcane",
    "Sunflower": "sunflower",
    "Tomato": "tomato",
    "Wheat": "wheat1"
}

# Load the models and label encoders
def load_model_and_encoders(file_name):
    with open(f"models/{file_name}_min_price_model.pkl", 'rb') as f:
        min_price_model = pickle.load(f)
    with open(f"models/{file_name}_max_price_model.pkl", 'rb') as f:
        max_price_model = pickle.load(f)
    with open(f"models/{file_name}_modal_price_model.pkl", 'rb') as f:
        modal_price_model = pickle.load(f)
    
    with open(f"models/{file_name}_label_encoders.pkl", 'rb') as f:
        label_encoders = pickle.load(f)
    
    return min_price_model, max_price_model, modal_price_model, label_encoders

# Function to encode input with unknown label handling
def encode_with_unknown(le, value):
    try:
        return le.transform([value])[0]
    except ValueError:
        return -1  # Assign -1 or some placeholder value for unseen labels

# The main function to predict prices based on commodity
def predict_prices(commodity, district, market_name, day, month, year):
    try:
        # Get the corresponding base file name for the commodity
        file_base = commodity_to_file_base.get(commodity)
        if not file_base:
            raise ValueError("Commodity not found in mapping.")
        
        # Load the corresponding model and encoders
        min_price_model, max_price_model, modal_price_model, label_encoders = load_model_and_encoders(file_base)

        # Encode the input data
        encoded_data = {
            'District Name': encode_with_unknown(label_encoders['District Name'], district),
            'Commodity': encode_with_unknown(label_encoders['Commodity'], commodity),
            'Market Name': encode_with_unknown(label_encoders['Market Name'], market_name),
            'Day': day,
            'Month': month,
            'Year': year,
        }

        # Prepare input features for prediction
        X_input = np.array([list(encoded_data.values())]).reshape(1, -1)

        # Make predictions
        min_price_pred = min_price_model.predict(X_input)
        max_price_pred = max_price_model.predict(X_input)
        modal_price_pred = modal_price_model.predict(X_input)

        # Return the predictions
        return {
            "Min Price": min_price_pred[0],
            "Max Price": max_price_pred[0],
            "Modal Price": modal_price_pred[0]
        }
    except Exception as e:
        return {"Error": str(e)}

# Example usage:
commodity = "Tomato"
district = "Chittor"
market_name = "Madanapalle"
day = 30
month = 10
year = 2029

predictions = predict_prices(commodity, district, market_name, day, month, year)
print(predictions)
```

```python
import pandas as pd
import pickle
import os
from sklearn.preprocessing import LabelEncoder

# Define the model directory
model_dir = "models1"

# Define a function to load the models and label encoders
def load_models_and_encoders(model_name):
    try:
        # Load the models
        with open(os.path.join(model_dir, f"{model_name}_min_price_model.pkl"), 'rb') as f:
            min_price_model = pickle.load(f)
        with open(os.path.join(model_dir, f"{model_name}_max_price_model.pkl"), 'rb') as f:
            max_price_model = pickle.load(f)
        with open(os.path.join(model_dir, f"{model_name}_modal_price_model.pkl"), 'rb') as f:
            modal_price_model = pickle.load(f)

        # Load the label encoders
        with open(os.path.join(model_dir, f"{model_name}_label_encoders.pkl"), 'rb') as f:
            label_encoders = pickle.load(f)

        print(f"Models and label encoders loaded for {model_name}")
        return min_price_model, max_price_model, modal_price_model, label_encoders

    except Exception as e:
        print(f"Error loading models for {model_name}: {e}")
        return None, None, None, None

# Define a function for making predictions
def make_predictions(new_data, model_name):
    # Load the models and label encoders
    min_price_model, max_price_model, modal_price_model, label_encoders = load_models_and_encoders(model_name)

    if min_price_model is None:
        print("Error loading models. Cannot make predictions.")
        return

    # Preprocess the new data
    # Here, we assume 'new_data' is a DataFrame with the same structure as the training data
    # Ensure the same encoding for categorical features
    for col in ['District Name', 'Commodity', 'Market Name']:
        if col in new_data.columns:
            encoder = label_encoders.get(col)
            if encoder:
                new_data[col] = encoder.transform(new_data[col])
            else:
                print(f"Warning: Encoder not found for column {col}")

    # Define the features for prediction (same as during training)
    X_new = new_data[['District Name', 'Commodity', 'Market Name', 'Day', 'Month', 'Year']]

    # Make predictions using the models
    min_price_pred = min_price_model.predict(X_new)
    max_price_pred = max_price_model.predict(X_new)
    modal_price_pred = modal_price_model.predict(X_new)

    # Return the predictions
    return min_price_pred, max_price_pred, modal_price_pred

# Example usage

# Example new data (replace this with actual data you want to predict)
new_data = pd.DataFrame({
    'District Name': ['Agra'],
    'Commodity': ['Bajra(Pearl Millet/Cumbu)'],
    'Market Name': ['Agra'],
    'Day': [10],
    'Month': [9],
    'Year': [2026]
})

# Specify the model name (use the same name as the file base name)
model_name = 'bajra'  # Replace with the actual model's base name (e.g., bajra)

# Make predictions
min_price_pred, max_price_pred, modal_price_pred = make_predictions(new_data, model_name)

print(f"Predicted Min Price: {min_price_pred}")
print(f"Predicted Max Price: {max_price_pred}")
print(f"Predicted Modal Price: {modal_price_pred}")
```
